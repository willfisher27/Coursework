```python
import pandas as pd
storms = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/dplyr/storms.csv")
```

```python
for col in storms.columns:
    print(col, storms[col].dtype)
```

```python
names = storms['name'].unique().tolist()
for i in names:
    df = storms[storms['name']==i]
    if not df.isnull().values.any():
        print(i)
```

```python
paloma = storms[storms['name']=='Paloma']
```

```python
paloma.to_csv('paloma.csv', index=False)
```

```python
paloma
```

```python
import matplotlib.pyplot as plt
colors = ['orange', 'pink', 'red', 'maroon']
stats = ['tropical depression', 'tropical storm', 'hurricane']
for s in stats:
    df = paloma[paloma['status']==s]
    if s=='hurricane':
        for num in range(4):
            df2 = df[df['category']==num+1]
            plt.scatter(df2['wind'], df2['pressure'], color=colors[num], label='Category '+str(num+1)+' Hurricane')
    elif s=='tropical storm':
        plt.scatter(df['wind'], df['pressure'], color='gold', label='Tropical Storm (Category 0)')
    else:
        plt.scatter(df['wind'], df['pressure'], color='green', label='Tropical Depression (Category -1)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Maximum Windspeed (knots)', fontsize=13)
plt.ylabel('Air Pressure at Center (millibars)', fontsize=13)
plt.title('Hurricane Paloma Pressure Versus Windspeed', fontsize=14)
plt.show()
```

```python
import numpy as np
x = paloma['long']
y = paloma['lat']
plt.figure(figsize=(10,10))
plt.scatter(x, y, s=np.array(paloma['tropicalstorm_force_diameter'])*75, label='Tropical storm diameter')
plt.scatter(x, y, s=np.array(paloma['hurricane_force_diameter'])*75, label='Hurricane diameter')
plt.scatter(x, y, label='Location of storm center')
plt.axis('equal')
plt.xticks([-83, -82, -81, -80, -79, -78, -77])
plt.xlabel('Longitude (Â°E)', fontsize=14)
plt.ylabel('Latitude (Â°N)', fontsize=14)
plt.legend(fontsize=16, loc=2)
plt.grid(True)
plt.title('Path and Size of Hurricane Paloma', fontsize=18)
plt.show()
```

```python
emp_data = pd.read_csv('Current_Employee_Names__Salaries__and_Position_Titles.csv')
```

```python
emp_data
```

```python
salaried = emp_data[emp_data['Salary or Hourly']=='Salary']
deps = salaried['Department'].unique().tolist()
del deps[len(deps)-2]
paid = {}
for dep in deps:
    df = salaried[salaried['Department']==dep]
    paid[dep] = df['Annual Salary'].mean()
paid = sorted(paid.items(), key=lambda x: x[1], reverse=True)
x = [i[0] for i in paid]
y = [int(i[1]) for i in paid]

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,7), constrained_layout=True)
plt.xlabel('Department', loc='left', fontsize=14)
plt.title('Highest and Lowest Average Chicago Government Salaries by Department')
ax1.bar(x[:10], height=y[:10])
ax1.set_xticklabels(x[:10], rotation='vertical')
ax1.set_title('10 Highest Average Departmental Salaries', fontsize=14)
ax1.set_ylabel('Salary ($)', fontsize=14)
ax2.bar(x[-10:], height=y[-10:])
ax2.set_xticklabels(x[-10:], rotation='vertical')
ax2.set_title('10 Lowest Average Departmental Salaries', fontsize=14)
plt.show()
```
