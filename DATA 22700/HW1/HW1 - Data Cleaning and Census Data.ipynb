```python
import pandas as pd
```

```python
#Problem1
cigs = pd.read_csv('Cigarette.csv')
```

```python
cigs
```

```python
cigs.drop(cigs.columns[0], axis=1, inplace=True)
df_lst = []
for i in range(1985, 1996):
    df = cigs[['state']+[col for col in cigs.columns if str(i) in col]]
    df.drop(df.columns[1], axis=1, inplace=True)
    df.columns = ['state']+[col[:-4] for col in df.columns[1:]]
    df['year'] = [i]*len(df)
    year =df.pop('year')
    df.insert(1, 'year', year)
    df_lst.append(df)
cigs_clean = pd.concat(df_lst, ignore_index=True)
cigs_clean
    
    
```

```python
cigs_clean.to_csv('cigs_clean', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = cigs_clean['year'].unique()
j = 0
for i in cigs_clean['state'].unique().tolist():
    st = cigs_clean[cigs_clean['state']==i]
    rgb = np.random.randint(3,)
    col = np.random.random(3)
    plt.scatter(x, st['packpc'], color=col)
    plt.plot(x, st['packpc'], color=col, label=i)
    j += 1
plt.xlabel('Year')
plt.ylabel('Packs Sold Per Capita')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Packs Sold Per Capita from 1985 to 1995')
plt.show()
```

```python
states = cigs_clean['state'].unique().tolist()
packd = {}
for i in states:
    df = cigs_clean[cigs_clean['state']==i]
    avg = df['packpc'].sum()/11
    packd[i] = avg
packds = sorted(packd.items(), key=lambda x: x[1])
packds
```

```python
#Problem2
census = pd.read_csv('census1900-2000.csv')
```

```python
census
```

```python
pop1900 = census[census['Year']==1900]['People'].sum()
pop2000 = census[census['Year']==2000]['People'].sum()
age_percent_1900 = {}
age_percent_2000 = {}
for i, r in census.iterrows():
    if r['Year'] == 1900:
        if r['Age'] in age_percent_1900:
            age_percent_1900[r['Age']] += r['People']/pop1900
        else:
            age_percent_1900[r['Age']] = r['People']/pop1900
    else:
        if r['Age'] in age_percent_2000:
            age_percent_2000[r['Age']] += r['People']/pop2000
        else:
            age_percent_2000[r['Age']] = r['People']/pop2000
age_percents_df = pd.DataFrame({'Age':census['Age'].unique().tolist(), 'Percent1900':list(age_percent_1900.values()), 'Percent2000':list(age_percent_2000.values())})
age_percents_df
```

```python
age = [0, 10, 20, 30, 40, 50, 60, 70, 80]
percent1900 = []
percent2000 = []
for j in age:
    a19 = 0
    a20 = 0
    for i, r in age_percents_df.iterrows():
        if r['Age'] >= j and r['Age'] < j + 10:
            a19 += r['Percent1900']
            a20 += r['Percent2000']
    percent1900.append(a19)
    percent2000.append(a20)
age_percents_df2 = pd.DataFrame({'Age':age, 'Percent1900':percent1900, 'Percent2000':percent2000})
age_percents_df2


    
```
